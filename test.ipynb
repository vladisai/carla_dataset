```python
import os

P = '/home/usobal/Repos/carla_dataset/outputs'

candidates = []

import os
for root, dirs, files in os.walk(P, topdown=False):
    for name in files:
        if name.endswith(".pkl"):
            candidates.append(os.path.join(root, name))
```

```python
print(len(candidates))
import random
```

```python
from collections import defaultdict, namedtuple


Box = namedtuple("Box", ["points", "distance_to_camera", "max_distance_to_center", "vehicle"])

import pickle
with open(random.choice(candidates), 'rb') as f:
    x = pickle.load(f)


from matplotlib import pyplot as plt

def draw(points):
    for i in range(4):
        j = (i + 1) % 4
        plt.plot([points[i][0], points[j][0]], [points[i][1], points[j][1]], c='r', linewidth=0.5)

def draw_d(d):    
    for b, q in zip(d['boxes'], d['boxes_q']):
        if q:
            draw(b.points)
    plt.imshow(d['image'])
    plt.axis('off')
    
plt.figure(dpi=500)
plt.subplot(3, 1, 1)
draw_d(x['FL'])
        
plt.subplot(3, 1, 2)
draw_d(x['FC'])

plt.subplot(3, 1, 3)
draw_d(x['FR'])
```
